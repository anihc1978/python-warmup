```python
import numpy as np

# Exercise 1: Create a 5x5 matrix of random numbers
rng = np.random.default_rng(42)
matrix = rng.random((5, 5))
print("Exercise 1: 5x5 random matrix")
print(matrix)
```

```python
# Exercise 2: Normalize each column to mean 0 / std 1 (z-score normalization)
col_means = matrix.mean(axis=0)
col_stds = matrix.std(axis=0)
normalized = (matrix - col_means) / col_stds
print("Exercise 2: Column-normalized matrix")
print(normalized)
print("\nColumn means (should be ~0):", normalized.mean(axis=0).round(10))
print("Column stds  (should be ~1):", normalized.std(axis=0).round(10))
```

```python
# Exercise 3: Compute pairwise Euclidean distances between rows
# Use broadcasting: ||a - b||^2 = ||a||^2 + ||b||^2 - 2*a.b
sq_sum = np.sum(matrix ** 2, axis=1, keepdims=True)  # (5,1)
dot_prod = matrix @ matrix.T                          # (5,5)
dist_matrix = np.sqrt(np.maximum(0, sq_sum + sq_sum.T - 2 * dot_prod))
print("Exercise 3: Pairwise Euclidean distance matrix (5x5)")
print(dist_matrix.round(4))
```

```python
# Exercise 4: Mask all values below the median (set them to NaN)
med = np.median(matrix)
masked = matrix.copy().astype(float)
masked[masked < med] = np.nan
print("Exercise 4: Matrix with below-median values masked (NaN)")
print(f"Median value: {med:.4f}")
print(masked)
```

```python
# Exercise 6: Reshape a 1D array of 100 values into 10x10
arr_1d = np.arange(100)
reshaped = arr_1d.reshape(10, 10)
print("Exercise 6: 1D array of 100 values reshaped to 10x10")
print(f"Original shape: {arr_1d.shape}")
print(f"New shape:      {reshaped.shape}")
print(reshaped)
```

```python
# Exercise 7: Compute a rolling mean with stride using stride_tricks
data = np.arange(1, 21, dtype=float)  # [1, 2, ..., 20]
window = 4
stride = 2

# Use stride_tricks to create sliding window view
from numpy.lib.stride_tricks import sliding_window_view
windows = sliding_window_view(data, window_shape=window)[::stride]
rolling_mean = windows.mean(axis=1)

print("Exercise 7: Rolling mean (window=4, stride=2)")
print(f"Input data:   {data}")
print(f"Rolling means: {rolling_mean}")
```

```python
# Exercise 8: One-hot encode an integer array
integer_array = np.array([0, 2, 1, 0, 3])
num_classes = integer_array.max() + 1
one_hot = np.eye(num_classes, dtype=int)[integer_array]
print("Exercise 8: One-hot encoding")
print(f"Integer array: {integer_array}")
print(f"Num classes:   {num_classes}")
print("One-hot encoded:")
print(one_hot)
```

```python
# Exercise 9: Concatenate two arrays along axis 0 then axis 1
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[5, 6], [7, 8]])

concat_axis0 = np.concatenate((arr1, arr2), axis=0)
concat_axis1 = np.concatenate((arr1, arr2), axis=1)

print("Exercise 9: Array concatenation")
print(f"Array 1:\n{arr1}")
print(f"Array 2:\n{arr2}")
print(f"\nConcatenated along axis 0 (rows):\n{concat_axis0}")
print(f"\nConcatenated along axis 1 (columns):\n{concat_axis1}")
```

```python
# Exercise 10: Replace all negative values with 0 (ReLU)
rng2 = np.random.default_rng(99)
array_with_neg = rng2.standard_normal((4, 5))  # values in ~[-3, 3]
relu_result = np.maximum(0, array_with_neg)
print("Exercise 10: Replace negative values with 0 (ReLU)")
print("Original array:")
print(array_with_neg.round(3))
print("\nAfter replacing negatives with 0:")
print(relu_result.round(3))
print(f"\nNegative values before: {(array_with_neg < 0).sum()}")
print(f"Negative values after:  {(relu_result < 0).sum()}")
```
